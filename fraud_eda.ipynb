```python
df['IsZeroAmount'] = (df['Amount'] == 0).astype(int)
df['LogAmount'] = np.log1p(df['Amount'])
```

```python
plt.hist(df['Amount'], bins=100)
plt.title("Raw Transaction Amount Distribution")
plt.show()
```

```python
plt.hist(np.log1p(df[df['Class']==0]['Amount']), bins=50, alpha=0.5, label='Legit')
plt.hist(np.log1p(df[df['Class']==1]['Amount']), bins=50, alpha=0.5, label='Fraud')
plt.legend()
plt.title("Log-Scaled Amount Distribution by Class")
plt.show()
```

```python
small_positive = df[(df['Amount'] > 0) & (df['Amount'] < 5)]
small_positive['Class'].mean()
```

```python
df['AmountBin'] = pd.qcut(df['Amount'], 10, duplicates='drop')
fraud_by_bin = df.groupby('AmountBin')['Class'].mean()
fraud_by_bin.plot(kind='bar')
plt.title("Fraud Rate by Amount Decile")
plt.show()
```

```python
df.groupby('AmountBin')['Class'].mean()
```

```python
df['Hour'] = (df['Time'] // 3600) % 24
```

```python
df['Hour'].value_counts().sort_index().plot()
plt.title("Transaction Volume by Hour")
plt.show()
```

```python
fraud_by_hour = df.groupby('Hour')['Class'].mean()
fraud_by_hour.plot()
plt.title("Fraud Rate by Hour")
plt.show()
```

```python
hourly_summary = (df.groupby('Hour').agg(total_transactions=('Class', 'count'),
          fraud_count=('Class', 'sum')))
hourly_summary['fraud_rate'] = (hourly_summary['fraud_count'] / hourly_summary['total_transactions'])

hourly_summary
```

```python
plt.figure()
hourly_summary['fraud_count'].plot()
plt.title("Fraud Count by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Count")
plt.show()
```

```python
df['IsNight'] = df['Hour'].between(0, 5).astype(int)
```

```python
df.groupby('IsNight')['Class'].mean()
```

```python
pivot_table = df.pivot_table(values='Class',index='Hour',columns='AmountBin',aggfunc='mean')
pivot_table = pivot_table.round(4)
#pivot_table
```

```python
count_pivot = df.pivot_table(values='Class',index='Hour',columns='AmountBin',aggfunc='count')
#count_pivot
```

```python
plt.figure(figsize=(18, 12))
plt.imshow(pivot_table, aspect='auto')
plt.colorbar(label='Fraud Rate')

for i in range(len(pivot_table.index)):
    for j in range(len(pivot_table.columns)):
        rate = pivot_table.iloc[i, j]
        count = count_pivot.iloc[i, j]
        plt.text(j, i, f"{rate:.2%}\n({count})",
                 ha='center', va='center', fontsize=6)

plt.xticks(range(len(pivot_table.columns)), pivot_table.columns, rotation=90)
plt.yticks(range(24))
plt.title("Fraud Rate with Transaction Count")
plt.show()
```

```python
sns.kdeplot(data=df[df['Class']==0], x='V17', label='Non-Fraud', fill=True)
sns.kdeplot(data=df[df['Class']==1], x='V17', label='Fraud', fill=True)
plt.legend()
plt.show()
```

```python
features = ['V14', 'V12', 'V10', 'V16']
plt.figure(figsize=(12, 10))

for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i) 
    sns.kdeplot(data=df[df['Class'] == 0],x=feature,label='Non-Fraud',fill=True)
    sns.kdeplot(data=df[df['Class'] == 1],x=feature,label='Fraud',fill=True)
    plt.title(f'{feature} Distribution by Class')
    plt.legend()

plt.tight_layout()
plt.show()
```
